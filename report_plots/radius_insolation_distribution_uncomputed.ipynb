# radius–insolation distribution without enrichment computations
- same plot as `radius_insolation_distribution.ipynb`, but on the merged-only table (`disable_calculations=True`),
  i.e. only values published by the catalogs/param sources, none derived by the pipeline
- axis limits fixed (as in the original notebook) since the auto range differs from `distr`
- exports to `report/figures/radius_insolation_distribution_uncomputed.pdf`

```python
%load_ext autoreload
%autoreload 2

import os, sys, pathlib

ROOT = pathlib.Path.cwd()
while not (ROOT / 'crossmatching').exists():
    ROOT = ROOT.parent
os.chdir(ROOT)
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import matplotlib.pyplot as plt
from astropy import units as u

from report_plots import common, data, plots
```

```python
cme, _ = data.emc_crossmatcher()
merger = data.build_param_filler()

u_emc = data.enrich(cme.catalog_table, cme, merger, disable_calculations=True)
u_emc["r_earth"] = u_emc["r"] * u.R_jupiter.to(u.R_earth)
len(u_emc)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plots.plot_distr(u_emc, ax)
plt.ylim(0.5E-3, 0.5E6)
plt.xlim(0.2E0, 1E2)
common.save_figure("radius_insolation_distribution_uncomputed")
```
