```python

def e2e_crossmatch(query_table):
    cm = Crossmatcher(NEACatalog(), SimbadIdSupplier())
    cm.load_catalog(from_file="../pscomppars.txt")
    # cm.load_alternate_ids(query_table["star_name"], from_file="../alternate_ids.txt")
    return cm.combined_crossmatch(query_table)
```

```python
# table = Table.read("./data/guinea_tics.csv", format="csv")
# table.write("./data/guinea_tics.txt", format="ascii", overwrite=True)  # Ensure file is in correct format
table = Table.read("./data/guinea_tics.txt", format="ascii")

query_table = table[["star_name", "ra", "dec", "sy_dist"]]

results = e2e_crossmatch(query_table)
results
```

```python

# Check that all expected planets are found
for star_name, expected_planets in table[["star_name", "expected_planets"]]:
    expected_planet_list = [p for p in expected_planets.split("|")]
    matched_planets = results[results["star_name"] == star_name]["pl_name"].tolist()
    for planet in expected_planet_list:
        # assert planet in matched_planets, f"Expected planet {planet} not found for {star_name}"
        if planet not in matched_planets:
            print(f"Expected planet {planet} not found for {star_name}")

# Check for no false positives (i.e. all matched planets should be in expected list)
for star_name in table["star_name"]:
    expected_planets = [p for p in table[table["star_name"] == star_name]["expected_planets"][0].split("|")]
    matched_planets = results[results["star_name"] == star_name]["pl_name"].tolist()
    for planet in matched_planets:
        assert planet in expected_planets, f"Unexpected planet {planet} found for {star_name}"

```

```python
tau_tic = "TIC 419015728"
tau_ceti = query_table[query_table["star_name"] == tau_tic]
tau_ceti_match = e2e_crossmatch(tau_ceti)

expected = table[table["star_name"] == tau_tic]["expected_planets"][0].split("|")
matched = tau_ceti_match["pl_name"].tolist()
print(expected)
print(matched)
```

```python
from astropy.coordinates import SkyCoord
from astropy.table import Table
import astropy.units as u

# HPIC row — tau_ceti already loaded above
tic_ra   = float(tau_ceti["ra"][0])
tic_dec  = float(tau_ceti["dec"][0])
tic_dist = float(tau_ceti["sy_dist"][0])
print(f"HPIC  TIC 419015728 → ra={tic_ra}, dec={tic_dec}, dist={tic_dist} pc")

# pscomppars row
psc = Table.read("../pscomppars.txt", format="ascii")
tau = psc[["tau Cet" in str(row["hostname"]) for row in psc]]
tau1 = tau[0]
psc_ra   = float(tau1["ra"])
psc_dec  = float(tau1["dec"])
psc_dist = float(tau1["sy_dist"])
print(f"pscomppars tau Cet  → ra={psc_ra}, dec={psc_dec}, dist={psc_dist} pc")

# 2D angular separation
c1 = SkyCoord(ra=tic_ra*u.deg, dec=tic_dec*u.deg, distance=tic_dist*u.pc)
c2 = SkyCoord(ra=psc_ra*u.deg, dec=psc_dec*u.deg, distance=psc_dist*u.pc)

sep_2d = c1.separation(c2)
sep_3d = c1.separation_3d(c2)

print(f"\n2D angular separation : {sep_2d.to(u.arcsec):.6f}")
print(f"                      : {sep_2d.deg:.8f} deg")
print(f"\n3D spatial separation : {sep_3d.to(u.pc):.6f}")
```

```python
# Debug: Check types of tables being vstacked
cm = Crossmatcher(NEACatalog(), SimbadIdSupplier())

# Run the individual steps
cm.id_crossmatch(query_table)
cm.coordinate_crossmatch(query_table)

# Get the filtered tables that would be vstacked
uuid = cm.planet_uid
matched_listid = cm.id_matched[uuid].tolist()
matched_list3d = cm.coords3d_matched[uuid].tolist()
matched_list2d = cm.coords2d_matched[uuid].tolist()

only_coords3d = cm.coords3d_matched[~np.isin(matched_list3d, matched_listid)]
only_coords2d = cm.coords2d_matched[
    ~(np.isin(matched_list2d, matched_listid) | np.isin(matched_list2d, only_coords3d[uuid].tolist()))
]

print(f"id_matched objectid type: {cm.id_matched['objectid'].dtype}")
print(f"only_coords3d objectid type: {only_coords3d['objectid'].dtype if len(only_coords3d) > 0 else 'empty'}")
print(f"only_coords2d objectid type: {only_coords2d['objectid'].dtype if len(only_coords2d) > 0 else 'empty'}")

# Check if objectid column exists in all tables
print(f"id_matched has objectid: {'objectid' in cm.id_matched.colnames}")
print(f"only_coords3d has objectid: {'objectid' in only_coords3d.colnames}")
print(f"only_coords2d has objectid: {'objectid' in only_coords2d.colnames}")
```
